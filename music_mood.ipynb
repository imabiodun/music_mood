```python
def get_metric(model,y_test,y_pred):
    """create a series frame showing the accuracy, precision, f1 and recall scores"""
    metric ={}
    metric ['accuracy'] = accuracy_score(y_test,y_pred)
    precision,recall,f_score,support = precision_recall_fscore_support(y_test,y_pred,average='weighted')
    metric['precision'], metric ['recall'] ,metric ['f1']= precision,recall,f_score
    metric ['train score'] = model.score(X_train,y_train) 
    return pd.Series(metric)
```

```python
pd.concat([get_metric(lr,y_test,lr.predict(X_test)),
           get_metric(lr_cv_1,y_test,lr_cv_1.predict(X_test)),
           get_metric(lr_cv_2,y_test,lr_cv_2.predict(X_test))],
          axis = 1).rename(columns = {0:'lr',1:'l1',2:'l2'})
```

```python
model_names = ['LR','LR_CV_1','LR_CV_2']
models = [lr,lr_cv_1,lr_cv_2]
```

```python
coeff_list = []
for index,mod in (zip(model_names,models)):
    multi_index = pd.MultiIndex.from_product([[index],[0,1,2,3]])
    coeff_list.append(pd.DataFrame(mod.coef_, index = multi_index).T)
coef_df = pd.concat(coeff_list,axis =1)
```

```python
fig,axes = plt.subplots(nrows= 2, ncols = 2, sharex= True ,figsize = (10,10))
# plt.subplot_mosaic("""AAAAAAAABBCCBBCC""")
axes = axes.flatten()
for index,frame in enumerate(axes):
    data = coef_df.xs(index,level = 1,axis =1)
    data.plot(marker = '.',ls='', ax=frame, legend = False)
    frame.set_title(f'Coefficient of {le.classes_[index]}')
    plt.suptitle('Coefficients of Models')
plt.legend(loc = 3)
plt.tight_layout()
plt.show()
```

```python
confu_mat = {}
for label,mod in zip(model_names,models):
    confu_mat[label] = confusion_matrix(y_test, mod.predict(X_test))
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,8))
axes = axes.flatten()
axes[-1].axis('off')
fig.set_size_inches(12, 10)

for ax,name in zip(axes, confu_mat):
    sns.heatmap(confu_mat[name], ax=ax, annot=True, fmt='d', xticklabels = le.classes_, yticklabels = le.classes_, cmap = 'winter');
    ax.set(title=f'Confusion Matrix of {name}');
plt.xlabel('True')    
plt.ylabel('Predicted')  
plt.tight_layout()
```
